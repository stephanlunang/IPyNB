```python
%pylab inline
from numpy import *
from matplotlib.pyplot import *
from scipy.optimize import leastsq
pi = 3.14159265359 
G = 6.67384*10**-11
c = 299792.458 #km/s
z,mu,sigma = np.loadtxt('SN.dat', usecols=(1,2,3), unpack=True)
```

```python
#print len(sigma), len(mu), len(z)
from scipy.integrate import quad
q = 0
i = 0

Om1 = .23
Om2 = 1
h = 0.7
H0 = 70 #km/s/Mpc
Ol1 = 1-Om1
Ol2 = 1-Om2
#Ez = sqrt(Om1*(1+z)**3+(1-Om1))
Dh = 3000
d = zeros(200)
Dc = zeros(200)
Dc2 = zeros(200)
Dl = zeros(200)
Dl2 = zeros(200)
mu2 = zeros(200)
mu3 = zeros(200)
while (q < 2):
    def integrand(q):
        return 1/sqrt(Om1*(1+q)**3 + Ol1)
    def integrand2(q):
        return 1/sqrt(Om2*(1+q)**3 + Ol2)
    Q = quad(integrand, 0, q)
    Q2 = quad(integrand2, 0, q) 
    #print Q
    Dc[i] = Q[0]*Dh
    Dc2[i] = Q2[0]*Dh
    d[i] = q
    Dl[i] = (1 + q)*Dc[i]
    Dl2[i] = (1 + q)*Dc2[i]
    mu2[i] = 25 - 5*log10(.7) + 5*log10(Dl[i])
    mu3[i] = 25 - 5*log10(.7) + 5*log10(Dl2[i])
    if (i == 0):
        mu2[i] = mu2[i] + .000000000000000000001 #~0, I put this in so I wouldnt get an error
        mu3[i] = mu3[i] + .000000000000000000001 #~0, I put this in so I wouldnt get an error
    i = i + 1
    q = q + .01
figure(figsize=(10,10))
#print mu2, '\n', mu3
plot(d, mu2, 'r', label = '$\Omega_m$ = .23')
plot(d, mu3,'g', label = '$\Omega_m$ = 1')
errorbar(z, mu, sigma, fmt = 'c.')
title('Hubble Diagram')
xlabel('Redshift, "z"')
ylabel('$\mu$')
legend()

#%load_ext sympyprinting
#import sympy as sym
#from sympy import *

#print 'The'
#print '$$ \Omega_m = .23$$' 
#fits the data better than the $$\Omega_m = 1$$ case.'
```
